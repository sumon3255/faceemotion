```python
import cv2
```

```python
from deepface import DeepFace
```

```python
img = cv2.imread('happyboy.jpg')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img) #BGR
```

```python
plt.imshow(cv2.cvtColor(img , cv2.COLOR_BGR2RGB)) #BGR
#imread method loads our image as a multi-dimensional NumPy array 
(one for each Red, Green, and Blue component, respectively) and imshow displays our image to our screen.
```

```python
predictions = DeepFace.analyze(img)
```

```python
predictions
```

```python
type(predictions)
```

```python
predictions['dominant_emotion']
```

# We are trying to draw rectangle at face

```python
#for image pass in deep we pass whole image
fascecascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")  #harcascade is face recognation algoritm
#boundary rectangles for the detected faces or eyes
```

```python
#converting gray
gray = cv2.cvtColor(img, cv2.COLOR_BGRA2GRAY)
faces = fascecascade.detectMultiScale(gray,1.1,4) #slect multiple scale 
#and draw a rectangle
for(x,y,w,h) in faces:
    cv2.rectangle(img,(x,y), (x+w,y+h),(0,255,0),2)
```

```python
plt.imshow(cv2.cvtColor(img , cv2.COLOR_BGR2RGB))
```

```python
font =cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
#using put text method
#for inserting put text on video
cv2.putText(img,predictions['dominant_emotion'],(0,50),font,1,(0,0,255),2,cv2.LINE_4)
```

```python
plt.imshow(cv2.cvtColor(img , cv2.COLOR_BGR2RGB))
```
